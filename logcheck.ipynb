```python
import csv
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dla_cpus = []
dla_gpu = []
dla_power = []
dla_on_gpu_power = []
dla_time = []

for i, r in enumerate(csv.reader(open('./logs/dla_cpu.csv', encoding='utf-8'))):
    if i == 0:
        continue
    cpus = [eval(str(ll))['user'] for ll in eval(r[1])]
    dla_cpus.append(sum(cpus) / len(cpus))


for i, r in enumerate(csv.reader(open('./logs/dla_gpu.csv', encoding='utf-8'))):
    if i == 0:
        continue
    gpus = eval(r[0])['status']['load']
    dla_gpu.append(gpus)


for i, r in enumerate(csv.reader(open('./logs/dla_power.csv', encoding='utf-8'))):
    if i == 0:
        continue
    dla = eval(r[0])['CV']['curr']
    gpu = eval(r[0])['GPU']['curr']
    dla_power.append(dla / 3)
    dla_on_gpu_power.append(gpu / 5)

for i, r in enumerate(csv.reader(open('./logs/dla_time.csv', encoding='utf-8'))):
    if i == 0:
        continue
    gpus = eval(r[0])['status']['load']
    dla_gpu.append(gpus)
```

```python
print(len(dla_cpus), len(dla_gpu), len(dla_power), len(dla_on_gpu_power))
dla_df = pd.DataFrame(list(zip(dla_cpus, dla_gpu, dla_power, dla_on_gpu_power)), columns=['cpus', 'gpus', 'dla_power', 'gpu_power'])
dla_df = dla_df[dla_df['dla_power'] > 0]
plt.ylim(0, 200)
plt.plot(dla_df['cpus'], label='cpu_utilization')
plt.plot(dla_df['gpus'], label='gpu_utilization')
plt.plot(dla_df['dla_power'], label='dla_watt')
plt.plot(dla_df['gpu_power'], label='gpu_watt')
plt.legend()
plt.show()
```
